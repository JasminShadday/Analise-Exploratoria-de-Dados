```python
'''
Dado os pares de pontos determine o coeficiente de
correlação e a equação do 1° grau que melhor modela
os pontos usando a técnica da regressão linear
a) (3,7) (2,5) (−1,−1) (4,9) (8,20)
b) (1,1) (2,3) (−1,2) (0,−1)
c) (−2,−9) (2,1) (4,6) (6,11) (0, −4)
'''

import numpy as np
import matplotlib.pyplot as plt

#a)
#xi = [3,2,-1,4,8]
#yi = [7,5,-1,9,20]

#b)
#xi = [1,2,-1,0]
#yi = [1,3,2,-1]

#c)
#xi = [-2,2,4,6,0]
#yi = [-9,1,6,11,-4]

xi = [1965,1970,1975,1980,1985,1990,1995,2000,2005,2010]
yi = [40,36,37,34,33,30,32,24,25,20]

n = len(xi)

x2 = []
y2 = []
xy = []

for i in range(n):
  x2.append(xi[i]**2)
  y2.append(yi[i]**2)
  xy.append(xi[i]*yi[i])

somax = sum(xi)
somay = sum(yi)
somax2 = sum(x2)
somay2 = sum(y2)
somaxy = sum(xy)

#calculo da correlação linear
rxy = (n*somaxy - somax*somay)/(np.sqrt(n*somax2 - somax**2)*
                                np.sqrt(n*somay2 - somay**2))

a = (n*somaxy - somax*somay)/(n*somax2 - somax**2)
b = (somay - a*somax)/n

#reta de regressão linear => y = a.x + b
x = np.linspace(min(xi),max(xi),100)
y = a*x + b
print(f"o valor do rxy = {rxy:3f}")
print(f"o valor de a = {a:3f}")
print(f"o valor de b = {b:3f}")
#plt.plot(x,y)
#plt.scatter(xi,yi, color = "red")
#plt.show()

#a) y = a.x+b
ra = a*2015 + b
print(f"a porcentagem de fumantes em 20215 é {ra:3f}")

#b)quero o ano que a porcentagem de fumantes seja zero
# x = -b/a
ano = -b/a
print(f"o ano que a porcentagem de fumantes é zero é {ano:3f}")
plt.plot(x,y)
plt.scatter(xi,yi, color = "red")
plt.show()
```
