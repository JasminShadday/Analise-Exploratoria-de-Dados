Gráfico de linhas


```python
#temperatura máxima e mínima em um mês

import matplotlib.pyplot as plt

eixo_x_data = [1, 12, 13, 14, 16, 26, 27, 30]
eixo_y_temmin = [15, 12, 10, 17, 11, 21, 19, 18]
eixo_y_temmax = [25, 32, 31, 30, 35, 34, 29, 36]

#adicionando título e legenda
plt.title("Temperatura Máxima e Mínima")
plt.xlabel("Dias")
plt.ylabel("Temperatura (C)")
plt.savefig("gráfico de linhas.png")
plt.grid()
plt.ylim([10, 40])
plt.legend(loc = "upper left")

#cor de fundo
#plt.figure(facecolor = "#00FFFF")

#Estilizando
plt.plot(eixo_x_data, eixo_y_temmin, marker="o", color="#00BFFF", linestyle="-", linewidth=1, label="Temperatura Minima")
plt.plot(eixo_x_data, eixo_y_temmax, marker="o", color="#B22222", linestyle="-", linewidth=1, label="Temperatura Máxima")

plt.show()
```
