```python
#gráfico de setores
import matplotlib.pyplot as plt
veiculos = ["ela","elo","cachorro","??","ele"]
quantidade = [30,12,5,15,8]
cores = ["#ffbf00","#1a0812","#f793eb","#855294","#006b83"]
plt.title("Pronomes em uma sala",
          color = "#1a0812",
          fontsize = 21)
plt.xlabel("Pessoas", fontsize = 12)
plt.ylabel("Quantidade", fontsize = 12)
plt.bar(veiculos,quantidade,color = cores)
for i, valores in enumerate(quantidade):
  plt.text(i,valores,str(valores), ha="center", va="bottom")
#plt.ylim([0,100])
plt.yticks(range(0,32,2))
plt.grid(axis="y", linestyle = "--", linewidth = 0.5)
plt.xticks(rotation =45)
plt.savefig("gráfico de colunas - pronomes.png")


plt.show()
```

```python
#orientação sexual
veiculos = ["bi","hetero","não binário","gay","sapatão"]
quantidade = [30,12,5,15,8]
cores = ["#ffbf00","#1a0812","#f793eb","#855294","#006b83"]
plt.pie(quantidade, labels = veiculos, autopct = "%1.1f%%")
plt.title("Orientação sexual em uma sala",
          color = "#1a0812",
          fontsize = 21)
plt.savefig("gráfico de pizza - orientação sexual.png")
plt.show()
```

```python
#gráfico de setores
veiculos = ["sim","não","não sei"]
quantidade = [450,320,40]
cores = ["#ffbf00","#1a0812","#f793eb"]
plt.pie(quantidade, labels = veiculos, autopct = "%1.1f%%",
        colors = cores, startangle=90)
plt.savefig("gráfico de pizza.png")
plt.show()
```
